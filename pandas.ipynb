```python
scores = pd.Series([80, 70.5, 91, 87.5, 73], index = ["Rakesh", "Rajesh", "Cherry", "Sam", "Jim"])
print(scores)
```

```python
scores = pd.Series({
    "Rakesh" : 80,
    "Rajesh" : 70.5,
    "Cherry" : 91,
    "Sam" : 87.5,
    "Jim" : 73
})
print(scores)
print("\n")
scores.values # array([80. , 70.5, 91. , 87.5, 73. ])

scores.index # Index(['Rakesh', 'Rajesh', 'Cherry', 'Sam', 'Jim'], dtype='object')
```

```python
scores = pd.Series({
    "Rakesh" : 80,
    "Rajesh" : 70.5,
    "Cherry" : 91,
    "Sam" : 87.5,
    "Jim" : 73
})

print(scores.sum()) # Adds the values in array
print(scores.mean()) # cal mean of the values
print(scores.min()) # Gives minimum value of the list
print(scores.max()) # Gives maximum value of the list
```
